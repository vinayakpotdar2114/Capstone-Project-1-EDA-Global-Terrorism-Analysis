## 1. Write a program in Python to check whether an integer is Armstrong number or not.

**What is Armstrong Number?**

It is a number which is equal to the sum of cube of its digits.
For eg: 153, 370 etc.
Lets see it practically by calculating it,

Suppose I am taking 153 for an example 

First calculate the cube of its each digits

  1^3 = (1 * 1 * 1) = 1

  5^3 = (5 * 5 * 5) = 125

  3^3= (3 * 3 * 3) = 27

Now add the cube 

1+125+27 = 153

It means 153 is an Armstrong Number.

````python
# Write Your Code Here
inp= int(input())
#print(type(inp))
digits = [int(x) for x in str(inp)]
#print(digits)
c=0

for a in digits: # cubing and dding the numbers to get the sum
  a=a**3
  c=c+a

if c==inp: # comparing the sum of cubes with the input number
  print("The number is an armstrong number")
else:
  print("The given input is not an Armstrong Number")
````

**What is Fibonacci Series?**

A Fibonacci series is a series in which next number is a sum of previous two numbers.

 

For example : 0, 1, 1, 2, 3, 5, 8 ……

In Fibonacci Series, first number starts with 0 and second is with 1 and then its grow like,

0

1


 
0 + 1 = 1

1 + 1 = 2

1 + 2 = 3

2 + 3 = 5 and 

so on…

How our program will behave?
Its Logic is different from Fibonacci series program in c using iterative method.

Here we have a function named fibonacci() which will take a input and then return element. Each time it will call itself to calculate the elements of the series. 

Like if someone given 6 as a input then our program should return,

0, 1, 1, 2, 3, 5

````python
# Write Your Code Here
def fibonacci(n):
    if(n <= 1):
        return n
    else:
        return(fibonacci(n-1) + fibonacci(n-2))
n = int(input("Enter number of terms:"))
print("Fibonacci sequence:")
for i in range(n):
    print(fibonacci(i))

  # I could not write the code for recursion so I referenced a website.
````

## 7. How do you choose a machine learning algorithm for a problem statement? 
## Explain Briefly.

There is a process that can be followed in general--

1) Understand the Data -- If the data is labeled then we will use supervised learning, if not then unsupervised learning.

2) Process and Transform -- Once the data is processed for feature engineering components like null value analysis etc, we can go ahead and transform the data if needed.

3) Selection of ML algorithm -- Once we have the set of data that we are going to work with along with the type of problem statement we are facing, we can narrow down the algorithms that we can use. 

Now, there is no saying that a particular algorithm will perform radically better than the other in the bunch. So, we try and test multiple algorithms.

Now, based on the application we will try to select the perfromance metric and tune it accordingly.

eg. For cancer detection we will focus on reducing Fales Positives and False Negatives.

Depending on the results of performance metrics, we can then figure out the best performing algorithm.

## 8. What is Semi- Supervised Learning and why is it so important?


Semi Supervised Learning combines the best of both world i.e. Supervised and Unsupervised learning by using the small amount of labeled data and large amount of unlabeled data.

This approach helps in two ways--
1- The model can predict labeled data
2- The model can predict the future labels

Importance--- 

Considering the fact that, the real world data and problems wont be as streamlined as a kaggle project. The data wont always be in the best of shape.
It may not even be labelled. While working with such nuanced data, Semi Supervised Learning is happenes to be the sweet spot where we can use some labeled data and work with it, while having the high volume of data in terms of unlabeled data.
